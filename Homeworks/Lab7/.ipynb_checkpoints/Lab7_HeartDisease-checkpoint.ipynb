##### **Remove missing values**

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

##### **Binarization**
- We binarize the categorical attribute using one-hot encoding: this creates 'n' new columns for each of the values represented in feature, and those new columns contain binary values to encode presence or absence of the feature-value

```python
df2 = pd.get_dummies(df, columns = ['cp', 'restecg', 'slope', 'thal'])
new_attr_list = list(df2.columns)
new_attr_list.remove('target')
df2 = df2[new_attr_list]
df2.head()
```

```python
num_attribute = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'ca']
index_numeric = []

for col in num_attribute:
    index_numeric.append(df2.columns.get_loc(col))
```

### **Modeling**

##### **Split dataset**

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
X = np.array(df2.values)
Y = np.array(df['target'].values)

seed = 101
test_size = .3

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

##### **Normalization**
- We normalize numeric attribute using StandarScaler: this will 'mean center' the attribute such that it has a mean of 0 and  standard deviation of 1

$ x_{new} = \frac{x - mean(x)}{std(x)}$

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(x_train[:, index_numeric])
x_train[:, index_numeric] = scaler.transform(x_train[:, index_numeric])
x_test[:, index_numeric] = scaler.transform(x_test[:, index_numeric]) 
```

##### **Create the models**

```python
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
def make_confusion_matrix(cf, categories='auto', cbar=True, cmap='Blues', title=None):
    group_counts = ["{0:0.0f}\n".format(value) for value in cf.flatten()]

    box_labels = [f"{v1}".strip() for v1 in group_counts]
    box_labels = np.asarray(box_labels).reshape(cf.shape[0],cf.shape[1])

    sb.heatmap(cf, annot=box_labels, fmt="", cmap=cmap, cbar=cbar, xticklabels=categories, yticklabels=categories)

    
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
    if title:
        plt.title(title)
```

```python
model = DecisionTreeClassifier(criterion='entropy', random_state=seed)
model.fit(x_train, y_train)
predictions = model.predict(x_test)
```

```python
print('Accuracy:', accuracy_score(y_test, predictions))
```

```python
categories = ['No Heart \nDisease', 'Heart Disease']
make_confusion_matrix(confusion_matrix(y_test, predictions), categories=categories, cmap='binary')
```

```python
print('Classification report')
print(classification_report(y_test, predictions))
```

```python
model2 = DecisionTreeClassifier(criterion='gini', random_state=42)
model2.fit(x_train, y_train)
predictions2 = model2.predict(x_test)
```

```python
print('Accuracy:', accuracy_score(y_test, predictions2))
```

```python
categories = ['No Heart \nDisease', 'Heart Disease']
make_confusion_matrix(confusion_matrix(y_test, predictions2), categories=categories, cmap='binary')
```

```python
print('Classification report')
print(classification_report(y_test, predictions2))
```
